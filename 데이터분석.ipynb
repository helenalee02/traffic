```python
import pandas as pd
data=pd.read_csv('S-DoT_NATURE_2025.04.28-05.04.csv',encoding='cp949')
loc=pd.read_excel('서울시 도시데이터 센서(S-DoT) 환경정보 설치 위치정보.xlsx')
```

```python
data=merged[['측정시간','위도','경도','소음 최대(dB)', '소음 평균(dB)', '소음 최소(dB)']]
data.head()
```

```python
data=data.dropna()
data
```

```python
loca=data[['위도','경도']]
a=loca.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
import folium

# 중심 좌표를 데이터의 평균값으로 설정 (혹은 특정 좌표로 지정 가능)
center = [a['위도'].mean(), a['경도'].mean()]

# folium 지도 객체 생성
m = folium.Map(location=center, zoom_start=12)

# 각 행의 위도, 경도에 마커 추가
for idx, row in a.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        icon=folium.Icon(color='red')
        
    ).add_to(m)

# 지도 출력 (Jupyter Notebook 환경에서는 m만 입력해도 지도 표시)
m

# 만약 HTML로 저장하고 싶다면:
# m.save('my_map.html')
```

```python
data['측정시간']=pd.to_datetime(data['측정시간'].str.replace('_',' '))
```

```python
target_date = pd.to_datetime('2025-04-30')
filtered_df = data[data['측정시간'] > target_date]
```

```python
filtered_df
```

```python
data = filtered_df.groupby(['위도', '경도'], as_index=False).agg(소음평균dB=('소음 최대(dB)', 'mean'))
data = data.rename(columns={'소음평균dB': '소음 평균(dB)'})
data.head()
```

```python
node=pd.read_csv('서울시 자치구별 도보 네트워크 공간정보.csv',encoding='cp949')
node.head()
```

```python
node.columns
```

```python
node=node[(node['시군구명'] == '중구') | (node['시군구명'] == '종로구')]
node.shape
```

```python
node[['경도', '위도']] = node['노드 WKT'].str.extract(
    r'POINT\(([-\d\.]+) ([-\d\.]+)\)'
).astype(float)
```

```python
node.head()
```

```python
node.shape
```

```python
link=node[node['노드링크 유형']=='LINK']
link.shape
```

```python
node=node[node['노드링크 유형']=='NODE']
node.shape
```

```python
df_nodes=node
df_noise=data
df_noise.head()
```

```python
import pandas as pd
import networkx as nx
from geopy.distance import geodesic
import folium
```

```python
G = nx.Graph()
for idx, row in df_nodes.iterrows():
    G.add_node(row['노드 ID'], pos=(row['위도'], row['경도']))
```

```python
link.head()
```

```python
link[['경도', '위도']] = link['링크 WKT'].str.extract(
    r'LINESTRING\(([-\d\.]+) ([-\d\.]+)\)'
).astype(float)
link.head()
```

```python
link['경도'] = link['링크 WKT'].str.extract(r'LINESTRING\(([^ ]+) ').astype(float)
link['위도'] = link['링크 WKT'].str.extract(r'LINESTRING\([^ ]+ ([^,]+),').astype(float)
```

```python
link.head()
```

```python
link.columns
```

```python
links=link[['시작노드 ID', '종료노드 ID', '링크 길이']]
```

```python
def get_noise(lat, lon):
    min_dist = float('inf')
    noise_val = None
    for _, noise_row in df_noise.iterrows():
        dist = geodesic((lat, lon), (noise_row['위도'], noise_row['경도'])).meters
        if dist < min_dist:
            min_dist = dist
            noise_val = noise_row['소음 평균(dB)']
    return noise_val
```

```python
links.isna().sum()
```

```python
graph_nodes = list(G.nodes)

# 시작노드ID와 종료노드ID 모두 그래프에 있는 경우만 남김
valid_links = links[
    links['시작노드 ID'].isin(graph_nodes) & links['종료노드 ID'].isin(graph_nodes)
]
```

```python
valid_links.shape
```

```python
links= valid_links.values.tolist()
```

```python
def get_lat_lon(node_id):
    if node_id in G.nodes:
        return G.nodes[node_id]['pos']
    else:
        return None
node_id = 71387
lat, lon = get_lat_lon(node_id)
print(f"위도: {lat}, 경도: {lon}")
```

```python
condition = df_noise['위도'] > 90
df_noise.loc[condition, ['위도', '경도']] = df_noise.loc[condition, ['경도', '위도']].values
```

```python
invalid_nodes_s = []
invalid_nodes_e=[]
for start, end, dist in links:
    if start not in G.nodes:
        invalid_nodes_s.append(start)
    if end not in G.nodes:
        invalid_nodes_e.append(end)

print("존재하지 않는 노드 ID:", invalid_nodes_s)
print("존재하지 않는 노드 ID:", invalid_nodes_e)
```

```python
from tqdm import tqdm

noise_weight_factor = 0.5

for start, end, dist in tqdm(links, desc="링크 처리 중"):
    start_lat, start_lon = get_lat_lon(start)
    end_lat, end_lon = get_lat_lon(end)
    noise_start = get_noise(start_lat, start_lon)
    noise_end = get_noise(end_lat, end_lon)
    avg_noise = (noise_start + noise_end) / 2
    weight = dist*noise_weight_factor + noise_weight_factor * avg_noise
    G.add_edge(start, end, weight=weight, distance=dist, noise=avg_noise)
```

```python
import networkx as nx
import folium

# 두 가지 다른 가중치로 최단 경로 계산
path_weight = nx.dijkstra_path(G, start_node, end_node, weight='weight')  # 기본 가중치 경로
path_noise = nx.dijkstra_path(G, start_node, end_node, weight='noise')    # 소음 기준 경로
path_dist= nx.dijkstra_path(G, start_node, end_node, weight='distance')    # 거리 기준 경로

# 좌표 추출
path_weight_coords = [G.nodes[node]['pos'] for node in path_weight]
path_noise_coords = [G.nodes[node]['pos'] for node in path_noise]
path_dist_coords = [G.nodes[node]['pos'] for node in path_dist]


# 지도 생성
m = folium.Map(location=path_weight_coords[0], zoom_start=14)

# 두 경로를 서로 다른 색상으로 추가
folium.PolyLine(path_weight_coords, color='blue', weight=5, opacity=0.7, popup='Weight Path').add_to(m)
folium.PolyLine(path_noise_coords, color='red', weight=5, opacity=0.7, popup='Noise Path').add_to(m)
folium.PolyLine(path_dist_coords, color='green', weight=5, opacity=0.7, popup='Dist Path').add_to(m)


folium.Marker(path_weight_coords[0], popup='Start', icon=folium.Icon(color='green')).add_to(m)
folium.Marker(path_weight_coords[-1], popup='End', icon=folium.Icon(color='red')).add_to(m)

for idx, row in df_noise.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        icon=folium.Icon(color='blue'),
        popup=row['소음 평균(dB)']
    ).add_to(m)

m.save('multi_path_navigation.html')
m
```
